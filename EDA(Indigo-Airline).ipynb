```python
def drop_null(df):
    null_percentage=df.isnull().sum()/df.shape[0]*100
    columns_to_drop=[col for col in df.columns if null_percentage[col]>25.00]
    df.drop(columns= columns_to_drop,inplace=True)
    return df
```
